# **Linking Git Repository using Databricks Repos**

### Step 1: Linking Git Repository using Repos

Navigated to **Repos** from the Databricks left navigation panel.

Since no external GitHub or GitLab repository was available initially, a new repository was created directly inside Databricks using the Repos feature.

Databricks automatically initialized the repository named **Databricks-workflows** with a default **main** branch.

The repository is now successfully linked to the Databricks workspace and visible under:
Workspace → Repos → <user-email> → Databricks-workflows.

This confirms successful Git integration within Databricks using Repos.
